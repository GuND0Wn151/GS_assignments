```python
class BooksRating:
    
    def __init__(self,serialNo,rating=[]):
        self.serialNo=serialNo
        
        self.rating=rating
    
    def addRating(self,rating):
        self.rating.append(rating)
    
class Books(BooksRating):
    
    def __init__(self,ISBN,title,author,year_of_pub,publisher,image_s,image_l,image_m,serialNo=0,rating=[]):
        self.ISBN = ISBN
        self.title=title
        self.author=author
        self.year_of_pub=year_of_pub
        self.publisher=publisher
        self.image_s=image_s
        self.image_l=image_l
        self.image_m=image_m
        if rating ==[]:
            
            BooksRating.__init__(self,serialNo,rating)
        else:
            BooksRating.__init__(self,serialNo,[rating])
    
    def changeTitle(self,new_title):
        self.title=new_title
        
    def display(self):
        print("The details of the book are:")
        return f"Title: {self.title}, Author: {self.author}, year of publication: {self.year_of_pub}, publisher: {self.publisher}\n"

class User:
    def __init__(self,serialNo,address,areacode):
        self.serialNo=serialNo
        self.address=address
        self.areacode=areacode
        
    def changeUser(self,data):
        if isinstance(data,str):
            self.address=address
            print("Updated the address for the user")
        else:
            self.areacode=areacode
            print("Update the areacode for the user")
        

        
class Library:
    
    books_data=[]
    user_data=[]
    lend_data=[]
    
    
    
    @staticmethod
    def addBook(obj):
        Library.books_data.append(obj)
        
    @staticmethod
    def checkISBN(isbn):
        for i in Library.books_data:
            if i.ISBN == isbn:
                print("The isbn is present in the library")
                print(i.display())
                return i
        print("The book is not present in the library")
        return False
    
    @staticmethod
    def returnBookObj(isbn):
        for i in Library.books_data:
            if i.ISBN == isbn:
                return i
    
    @staticmethod
    def delBook(isbn):
        for i in range(len(Library.books_data)):
            if Library.books_data[i].ISBN==isbn:
                Library.books_data.pop(i)
                return
    
    @staticmethod
    def addUser(userObj):
        Library.user_data.append(userObj)
        
    @staticmethod
    def getUser(userid):
        for i in Library.user_data:
            if i.serialNo==userid:
                return i
        return None
        
    @staticmethod
    def lendBook(userid,bookid,issuedate):
        temp=issuedate.split("-")
        
        Library.lend_data.append([userid,bookid,date(int(temp[2]),int(temp[1]),int(temp[0]))])
        
    @staticmethod
    def returnBook(userid,bookid):
        tod=datetime.date.today()
        for i in range(len(Library.lend_data)):
            if Library.lend_data[i][0]==userid and Library.lend_data[i][1]==bookid:
                    popped=Library.lend_data.pop(i)
                    print("you have returned the book")
                    lendPeriod = abs((datetime.date.today()-popped[2]).days)
                    if lendPeriod>7:
                        print(f"The total lending cost is {(lendPeriod-7)*10}")
                    else:
                        print("No cost in lending, thank you.")
                        
        print("Done :)")
    
    @staticmethod
    def checktitle(title):
        for i in Library.books_data:
            if title==i.title:
                print("The book is present in the book library the details are: \n",Books.display(i))
                return
        print("The book title is not present in the book library")
       
    
    @staticmethod
    def avgRating(publisher):
        tempDic = dict()
        for i in Library.books_data:

            try:
                tempDic[i.publisher].append[sum(i.rating)/len(i.rating) if len(i.rating)!=0 else 0]
            except:
                tempDic[i.publisher]=[sum(i.rating)/len(i.rating) if len(i.rating)!=0 else 0]

                
        for i in tempDic.keys():
            if i==publisher:
                print(f"The Avg rating for the publisher {i} is {sum(tempDic[i])/len(tempDic[i])}")
            
    @staticmethod
    def total():
        tempDic = dict()
        for i in Library.books_data:
            try:
                tempDic[i.year_of_pub]+=1
            except:
                tempDic[i.year_of_pub]=1
        for i in tempDic:
            print(f"The no of books published in {i} year are {tempDic[i]}")

    @staticmethod
    def getCatalog():
        for i in Library.books_data:
            print(i.display())
    
    @staticmethod
    def topFive():
        tempList = []
        for i in Library.books_data:
            #print(i.rating)
            tempList.append([sum(i.rating)/len(i.rating),i.title,i])
        tempList.sort()
        
        for i in range(5):
            try:
                print(f"{i+1}. {tempList[i][2].display()} \n")
            except:
                print("Done")
                break
```

```python
def startManagement():
    while True:
        print("\n##################Book Management System##################\n")
        print("1. User Options\n2. Book Options\n3. Rating Options\n4. Exit")
        obj_option=input("Enter your options from the above list (Enter the no): ")


        if obj_option=="1":
            print("1. Add User\n2. Lend Book\n3. Return Book\n")
            user_input = input("Enter the option no: ")
            if user_input=="1":
                user_id=input("Enter the User Id: ")
                user_address = input("Enter the user Address: ")
                user_areacode = input("Enter the user areacode: ")
                new_userObj = User(user_id,user_address,user_areacode)
                Library.addUser(new_userObj)

            elif user_input=="2":
                user_id = input("Enter the User Id: ")
                book_id = input("Enter the book Id: ")
                lend_date = input("Enter the date in dd-mm-yyyy format: ")
                Library.lendBook(user_id,book_id,lend_date)
                print(f"\nThe book with id {book_id} is issued to user with id {user_id}")

            elif user_input == "3":
                user_id = input("Enter the User Id: ")
                book_id = input("Enter the book Id: ")
                Library.returnBook(user_id,book_id)

            else:
                print("please enter a valid no as shown above list.")


        elif obj_option=="2":
            print("1. Display books using ISBN\n2. Add book\n3. Delete book\n4. Get Top 5 Books\n5. Get Publisher Ratings\n6. Total Books Per Year\n7. Check Title\n8. View Catalog\n")
            user_input = input("Enter the option no: ")

            if user_input == "1":
                user_isbn = input("Enter the books isbn no: ")
                Library.checkISBN(user_isbn)

            elif user_input == "2":
                user_book_id = input("Enter the ISBN no: ")
                user_book_title = input("Enter the books title: ")
                user_book_author = input("Enter the books author name: ")
                user_book_publisher = input("Enter the books publisher name: ")
                user_book_year = input("Enter the year of publication of the book: ")
                user_book_image_s = input("Enter the small url of the book: ")
                user_book_image_l = input("Enter the large url of the book: ")
                user_book_image_m = input("Enter the medium url of the book: ")

                if user_book_id!="" and user_book_title!="" and user_book_author!="" and user_book_year!="" and user_book_publisher!="" and user_book_image_s!="" and user_book_image_l!="" and user_book_image_m!="":
                    bookobj = Books(user_book_id,user_book_title,user_book_author,user_book_year,user_book_publisher,user_book_image_s,user_book_image_l,user_book_image_m)
                    Library.addBook(bookobj)
                else:
                    print("please enter valid values for all the attributes")

            elif user_input == "3":
                user_isbn = input("Enter the isbn id which you want to delete: ")
                Library.delBook(user_isbn)

            elif user_input == "4":
                Library.topFive()

            elif user_input == "5":
                user_publisher = input("Enter the publisher which you want to see the rating")
                Library.avgRating(user_publisher)

            elif user_input == "6":
                Library.total()

            elif user_input == "7":
                user_title = input("Enter the title you want to search: ")
                Library.checktitle(user_title)
            elif user_input == "8":
                Library.getCatalog()
            else:
                print("please enter a valid input")

        elif obj_option=="3":
            print("1. Add rating")
            user_input=input("Enter the ISBN no: ")
            obj = Library.returnBookObj(user_input)
            user_rating=float(input("Enter the rating in a range of 1-10: "))
            if obj:
                obj.addRating(user_rating)
                print("THe rating has been added to the book :)")
            else:
                print("No book found with the following ISBN no.")
        else:
            break
```
